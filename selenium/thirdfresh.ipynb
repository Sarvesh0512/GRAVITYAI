```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options  # Import Options
from webdriver_manager.chrome import ChromeDriverManager



# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (without GUI)
chrome_options.add_argument("--disable-gpu")  # Applicable for headless mode to disable GPU
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--start-maximized")  # Start maximized


# Install and set up the Chrome driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
# Open Google Maps
driver.get("https://www.google.com/maps")

# Check if the page title contains 'Google Maps' to ensure it has loaded
if "Google Maps" in driver.title:
    print("Google Maps opened successfully!")
else:
    print("Failed to open Google Maps.")

# Close the driver
driver.quit()
import logging
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
import time

# Set up logging configuration
logging.basicConfig(level=logging.DEBUG, filename="selenium_error.log",
                    format="%(asctime)s - %(levelname)s - %(message)s")

# Path to your chromedriver
chrome_driver_path = r"C:\Users\sarve\OneDrive\Desktop\WORK\selenium\chromedriver-win64\chromedriver.exe"

# Set up the Chrome driver using the Service class
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service)

def get_lat_lng(address):
    try:
        # Open Google Maps
        driver.get('https://www.google.com/maps')

        # Find the search box using its name attribute value
        search_box = driver.find_element("name", 'q')

        # Send the address to the search box
        search_box.send_keys(address)

        # Simulate hitting the Enter key
        search_box.send_keys(Keys.RETURN)

        # Allow time for the results to load
        time.sleep(5)

        # Extract the URL of the current page
        url = driver.current_url

        # The coordinates are usually in the URL
        if '@' in url:
            coords_part = url.split('@')[1].split(',')[0:2]
            lat = coords_part[0]
            lng = coords_part[1]
            return lat, lng
        else:
            return None, None
    except Exception as e:
        logging.error(f"Error fetching coordinates for {address}: {e}")
        return None, None

# Debugging: Print the current working directory
print(f"Current Working Directory: {os.getcwd()}")

# Absolute path to the Excel file
file_path = r"C:\Users\sarve\OneDrive\Desktop\WORK\selenium\1500to2000.xlsx" # Replace with the actual absolute path

# Ensure the file exists before attempting to read
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path}")

# Read the Excel file
df = pd.read_excel(file_path)

# Create columns for latitude and longitude
df['Latitude'] = None
df['Longitude'] = None

# Iterate over each row in the DataFrame and get the coordinates
for index, row in df.iterrows():
    address = row['Name']
    latitude, longitude = get_lat_lng(address)
    df.at[index, 'Latitude'] = latitude
    df.at[index, 'Longitude'] = longitude

# Save the updated DataFrame to a new Excel file
output_path = r"C:\Users\sarve\OneDrive\Desktop\WORK\selenium\updated1500to2000.0.xlsx"  # Replace with the actual absolute path
df.to_excel(output_path, index=False)

# Close the driver
driver.quit()

print("Latitude and longitude search completed.")
```
